```python
from statsmodels.tsa.stattools import adfuller
from numpy import log
result = adfuller(df.Rainfall.dropna())
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
```

```python
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
fig, axes = plt.subplots(3, 2, sharex=True)
axes[0, 0].plot(df.Rainfall); axes[0, 0].set_title('Original Series')
plot_acf(df.Rainfall, ax=axes[0, 1])

# 1st Differencing
axes[1, 0].plot(df.Rainfall.diff()); axes[1, 0].set_title('1st Order Differencing')
plot_acf(df.Rainfall.diff().dropna(), ax=axes[1, 1])

# 2nd Differencing
axes[2, 0].plot(df.Rainfall.diff().diff()); axes[2, 0].set_title('2nd Order Differencing')
plot_acf(df.Rainfall.diff().diff().dropna(), ax=axes[2, 1])

plt.show()
```
